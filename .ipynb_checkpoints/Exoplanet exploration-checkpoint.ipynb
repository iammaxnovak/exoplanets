```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#import the file
df = pd.read_csv("C:/Users/Max/Documents/Python/Data/planets.csv")

#Print out the top 5 entries, the shape and the columns names in planets, using display as looks nice
display(df.head())
display(df.shape)
display(df.columns)

#facilities that have found planets
print(df["pl_facility"].value_counts())
```

```python
#Sort the entries in descending order by orbit to see the largest

df_sort = df.sort_values("pl_orbper", ascending =False)

display(df_sort.head(1))
print(np.dtype(df_sort["pl_orbper"]))
```

```python
#Simple plot of the orbit and mass

fig , ax  = plt.subplots()
ax.axis([0.1,10000000,0.001, 100])
ax.loglog()

fig = plt.plot(df["pl_orbper"], df["pl_bmassj"], 'o', alpha= 0.5, markersize = 2)
fig = plt.xlabel("Planetry orbit (Earth days)")
fig = plt.ylabel("Planetry Mass (Jupiter Mass')")
plt.show()
```

```python
# List the discovery method

disc = df["pl_discmethod"].value_counts()
print(disc)
```

```python
#Create mapping dictionary to reduce the number of values, to tidy the graph a little

mapping = {"Radial Velocity":"Radial Velocity", "Imaging":"Imaging", "Eclipse Timing Variations": "Timing Variations", "Transit":"Transit", "Astrometry":"Astrometry", "Disk Kinematics":"Radial Velocity",
           "Orbital Brightness Modulation":"Orbital Brightness Modulation", "Pulsation Timing Variations":"Timing Variations", "Microlensing":"Microlensing", "Transit Timing Variations":"Timing Variations", "Pulsar Timing":"Pulsar Timing"}

df["mapping"] = df["pl_discmethod"].map(mapping)

#Graph showing planetry orbit against planetry mass on a logorithmic scale, color donotes discovery method

sns.set(style = "darkgrid", rc={'figure.figsize':(15,10)})
fig , ax  = plt.subplots()


g = sns.scatterplot(x="pl_orbper", y="pl_bmassj", data=df, alpha = 0.7, hue="mapping", s= 50, style = "mapping")
ax.set(xscale="log", yscale = "log")
ax.axis([0.1,10000000,0.001, 100])

legend = ax.legend()
legend.texts[0].set_text("Discovery Method")

plt.title("Confirmed exoplanet discoveries: Orbit in days vs Jupiter Mass")
plt.xlabel("Planetry orbit (days)")
plt.ylabel("Planetry mass (Jupiter mass')")
g.set_xticklabels(['0.01','0.1','1','10','100','1000','10e4','10e5', '10e6', '10e7'])
g.set_yticklabels(['0.0001','0.001','0.01','0.1','1','10', '100'])

#Add a pointer for Earth, using rough measurments"""

plt.scatter(365,0.091, marker ='o' , s=100, color = 'black', label = "Earth", alpha = 0.5)
plt.text(100+365,0.085, "Earth", horizontalalignment='left', size='medium', color='black', weight='semibold')


plt.savefig("Orbit V Mass")
plt.show()

```
